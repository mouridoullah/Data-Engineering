```python
def map(key,value):
    
    intermediate=[]
    for word in value.split():
        intermediate.append((word, 1))
    return intermediate

def reduce(key, values):
    
    result = 0
    for c in values:
        result = result +c
    return (key, result)

# Définir l'ensemble de lots
lots = {
    "./lot1.txt" : "jour lève notre grisaille",
    "./lot2.txt" : "trottoir notre ruelle notre tour",
    "./lot3.txt" : "jour lève notre envie vous",
    "./lot4.txt" : "faire comprendre tous notre tour"
}

# Appliquer la fonction map
intermediate_results = []
for key, value in lots.items():
    intermediate_results.extend(map(key, value))

# Regrouper les paires clé-valeur intermédiaires
intermediate_dict = {}
for key, value in intermediate_results:
    if key not in intermediate_dict:
        intermediate_dict[key] = []
    intermediate_dict[key].append(value)

# Appliquer la fonction reduce
final_results = []
for key, values in intermediate_dict.items():
    final_results.append(reduce(key, values))

# Afficher les résultats finaux
for result in final_results:
    print(result)
```

```python
import matplotlib.pyplot as plt

# Demander à l'utilisateur d'entrer un chiffre
n = int(input("Veuillez entrer un chiffre: "))

# Define the factorial function
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

# Define the range of x values
x_values = list(range(0, n+1))

# Calculate the corresponding y values
y_values = [factorial(x) for x in x_values]

# Create the plot
plt.plot(x_values, y_values, marker='o')

# Add title and labels
plt.title('Factorial Function')
plt.xlabel('n')
plt.ylabel('n!')

# Show the plot
plt.show()


# Demander à l'utilisateur d'entrer un chiffre
n = int(input("Veuillez entrer un chiffre: "))

# Define the range of x values
x_values = list(range(0, n+1))

# Calculate the corresponding y values
y_values = [x**2 for x in x_values]

# Create the plot
plt.plot(x_values, y_values, marker='o')

# Add title and labels
plt.title('Square Function')
plt.xlabel('n')
plt.ylabel('n^2')

# Show the plot
plt.show()
```
